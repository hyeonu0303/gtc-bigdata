```python
import pandas as pd

file_path = '/Users/hyeonu/Desktop/Code/bigdata-straysdog/src/python/excel/유기동물.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

df.head()
```

```python
# '자치구별(2)'과 '2022년' 컬럼을 추출
abandoned_animals = df[['자치구별(2)', 2022]].iloc[2:]
# 이름
abandoned_animals.rename(columns={'자치구별(2)': '자치구', 2022: '유기동물수'}, inplace=True)

# 결과 확인
abandoned_animals.sort_values(by="유기동물수",ascending=False,inplace=True)
abandoned_animals
```

```python
json_file_path = '../json/유기동물.json'
abandoned_animals.to_json(json_file_path, orient='records', force_ascii=False)

# # 저장된 파일 경로 반환
json_file_path
```
