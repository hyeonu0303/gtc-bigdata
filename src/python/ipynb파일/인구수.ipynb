```python
# 코드 실행 상태가 재설정되었으므로, 다시 필요한 라이브러리를 임포트하고 코드를 실행합니다.
import pandas as pd

# 엑셀 파일 불러오기
population_file_path = '../excel/인구밀도_20231130132635.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(population_file_path, header=0)
df.head()

```

```python
# 동별(2)에서 NaN이 아닌 행만 필터링
filtered_df = df[pd.notna(df.iloc[:, 1])]
filtered_df
```

```python
# 필요한 열만 선택
population = filtered_df.iloc[:, [1, 9]]
population
```

```python
# 첫 번째 행인 '소계' 부분을 제거
population = population.iloc[1:]
# 컬럼명 변경
population = population.rename(columns={'동별(2)': '자치구', 2022 : '인구수'})
#인구수 1000으로나눠줌
population['인구수'] = (population['인구수'] / 1000).astype(int)
population
```
