```python
#인구수 내림차순정렬
import pandas as pd
population_file_path = '../../excel/인구밀도_20231130132635.xlsx'
df = pd.read_excel(population_file_path, header=0)
filtered_df = df[pd.notna(df.iloc[:, 1])]
population = filtered_df.iloc[:, [1, 9]]
population = population.iloc[1:]
population = population.rename(columns={'동별(2)': '자치구', 2022 : '인구수'})
population['인구수'] = (population['인구수'] / 1000).astype(int)
population.sort_values(by='인구수',ascending=False,inplace=True)
population.head(10)
```

```python
#유기동물데이터
file_path = '../../excel/유기동물.xlsx'
df = pd.read_excel(file_path)
df.head()
abandoned_animals = df[['자치구별(2)', 2022]].iloc[4:]
abandoned_animals.rename(columns={'자치구별(2)': '자치구', 2022: '유기동물수'}, inplace=True)
abandoned_animals.head(10)
```

```python
#데이터합치기 '자치구' 기준
merged_df = pd.merge(population, abandoned_animals, on='자치구')

# 결과 출력
merged_df
```

```python
json_file_path = '../../json/인구수(내림)+유기동물.json'
merged_df.to_json(json_file_path, orient='records', force_ascii=False)

# # 저장된 파일 경로 반환
json_file_path
```
