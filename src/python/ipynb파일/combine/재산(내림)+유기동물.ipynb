```python
import pandas as pd
```

```python
file_path = '../../excel/지방세징수(구별)_20231203000721.xlsx'
df = pd.read_excel(file_path, header=3)
df.rename(columns={'Unnamed: 0': '자치구'}, inplace=True)
selected_columns = df[['자치구', '재산세']]
property_tax = selected_columns[selected_columns['자치구'] != '소계']
property_tax_copy = property_tax.copy()
property_tax_copy.sort_values(by='재산세', ascending=False, inplace=True)
property_tax_copy['재산세'] = (property_tax_copy['재산세'] / 1000).astype(int)
property_tax = property_tax_copy
print(property_tax)
```

```python
#유기동물데이터
file_path = '../../excel/유기동물.xlsx'
df = pd.read_excel(file_path)
df.head()
abandoned_animals = df[['자치구별(2)', 2022]].iloc[4:]
abandoned_animals.rename(columns={'자치구별(2)': '자치구', 2022: '유기동물수'}, inplace=True)
abandoned_animals.head(10)
```

```python
#데이터합치기 '자치구' 기준
merged_df = pd.merge(property_tax, abandoned_animals, on='자치구')

# 결과 출력
merged_df
```

```python
json_file_path = '../../json/재산세(내림)+유기동물.json'
merged_df.to_json(json_file_path, orient='records', force_ascii=False)

# # 저장된 파일 경로 반환
json_file_path
```
