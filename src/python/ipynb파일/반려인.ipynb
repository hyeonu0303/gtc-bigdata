```python
import pandas as pd
```

```python
file_path = '../csv/반려동물_유무_취득경로.csv'
df = pd.read_csv(file_path, header=2)
df
```

```python
selected_rows = df[df['구분별(1)'] == '지역소분류']
selected_rows.head()
```

```python
pet_df = selected_rows[['구분별(2)', '있다']]
pet_df.head()
```

```python
# pet_df의 복사본을 만들고 컬럼 이름 변경
pet_df_copy = pet_df.copy()
pet_df_copy.rename(columns={'구분별(2)': '자치구', '있다': '반려인비율'}, inplace=True)

# 변경된 데이터프레임 사용
# 예: 출력
pet_df_copy.head()
```

```python
pet_df_copy.sort_values(by="반려인비율", ascending=False, inplace=True)
pet_owner = pet_df_copy
pet_owner
```

```python
json_file_path = '../json/반려인.json'
pet_owner.to_json(json_file_path, orient='records', force_ascii=False)

# # 저장된 파일 경로 반환
json_file_path
```
