```python
import pandas as pd
```

```python
file_path = '../excel/유기동물보호+현황_20231204022829.xlsx'
df= pd.read_excel(file_path)
# 데이터프레임의 첫 몇 줄을 출력하여 구조 확인
df
```

```python
columns = {
    '시점': '년도',
    '합계': '개_총계',
    '합계.1': '개_인도(주인)',
    '합계.2': '개_입양분양',
    '합계.3': '개_폐사안락사',
    '합계.4': '개_계류기증',
    '합계.5': '고양이_총계',
    '합계.6': '고양이_인도(주인)',
    '합계.7': '고양이_입양분양',
    '합계.8': '고양이_폐사안락사',
    '합계.9': '고양이_계류기증'
}
# 첫 두 행(헤더 행)을 제거
df_cleaned_korean = df.drop([0, 1]).drop(columns=['자치구별(1)'])

# 열 이름 변경
df_cleaned_korean = df_cleaned_korean.rename(columns=columns)

# '년도' 열을 인덱스로 설정
df_cleaned_korean.set_index('년도', inplace=True)

df_cleaned_korean.head()
```

```python
# 열 합치기
df_cleaned_korean['개_입양분양_인도_계류기증'] = df_cleaned_korean['개_입양분양'] + df_cleaned_korean['개_계류기증'] + df_cleaned_korean['개_인도(주인)']
df_cleaned_korean['고양이_입양분양_인도_계류기증'] = df_cleaned_korean['고양이_입양분양'] + df_cleaned_korean['고양이_계류기증'] + df_cleaned_korean['고양이_인도(주인)']

# 원래 열 삭제
df_merged = df_cleaned_korean.drop(columns=['개_인도(주인)', '개_입양분양', '개_계류기증', '고양이_인도(주인)', '고양이_입양분양', '고양이_계류기증'])

df_merged.head()
```

```python
json_file_path = '../json/유기동물통계.json'
df_merged.to_json(json_file_path, orient='records', force_ascii=False)

# # 저장된 파일 경로 반환
json_file_path
```
