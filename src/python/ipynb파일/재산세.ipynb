```python
import pandas as pd
file_path = '../excel/지방세징수(구별)_20231203000721.xlsx'

```

```python
df = pd.read_excel(file_path, header=3)
# Unnamed:0을 -> 자치구로변경
df.rename(columns={'Unnamed: 0': '자치구'}, inplace=True)
df.head()
```

```python
# '자치구'와 '재산세' 컬럼 선택
selected_columns = df[['자치구', '재산세']]
selected_columns
```

```python
#자치구에서 소계 두줄 제거
property_tax = selected_columns[selected_columns['자치구'] != '소계']
#원본유지를위해 복사본생성
property_tax_copy = property_tax.copy()
#내림차순정렬
property_tax_copy.sort_values(by='재산세', ascending=False, inplace=True)
#재산세에 1000으로나눔
property_tax_copy['재산세'] = (property_tax_copy['재산세'] / 1000).astype(int)
property_tax = property_tax_copy
# 결과 확인
property_tax
```
